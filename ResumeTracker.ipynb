```python
import docx2txt
```

```python
# loading the resume & printing it
resume = docx2txt.process("ResumeElma.docx")
print(resume)
```

```python
job_des = docx2txt.process("JobDes.docx")
print(job_des)
```

```python
 #list of tesxt
text=[resume, job_des ]
```

```python
#counting the word occurnce in both resume
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print(cosine_similarity(count_matrix))
```

```python
match = cosine_similarity(count_matrix)[0][1]
match = match*100
match = round(match,2)
print(match)
```

```python
if(match > 70):
    print("Application is accepted")
else:
    print("Application is not selected")
```
